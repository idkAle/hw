```python
package = 99
quantity = int(input("How many would you like to purchase? \n"))
discount = 0
if quantity >= 10 and quantity <= 19:
  discount =.1
elif quantity >= 20 and quantity <= 49:
  discount =.2
elif quantity >= 50 and quantity <= 99:
  discount =.3
elif quantity >= 100:
  discount =.4

total_amount = package * quantity *(1 - discount)
print ("The discount saved is" , discount * 100, " %.")
print ("The total amount is ", total_amount)
```

```python

weight = int(input("What is the weight of your package? \n"))
rate = 0
if weight >= 1 and weight <= 2:
  rate = 1.50
elif weight > 2 and weight < 6:
  rate = 3.00
elif weight > 6 and weight < 10:
  rate = 4.00
elif weight > 10:
  rate = 4.75

total_amount = weight * rate
print ("The total amount is ", total_amount)
```
