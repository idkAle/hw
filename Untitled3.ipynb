```python
def func1(x,y):
  while y >= x:
    x = x + 1
    return x

num1 = int(input("First number"))
num2 = int(input("First number"))

sum = func1(num1,num2)
print(sum)
```

```python
def isPrime(num):
  for i in range(2,num):
    if(num % i == 0):
      return False
  return True

def listPrimes(x,y):
  for i in range(x,y+1):
    if (isPrime(i)):
      print(i, end = "  |  ")

listPrimes(2,1000)
```

```python
def coprime_test(num1,num2):
  for i in range(2,num2 + 1):
    if(num1 % i == 0 and num2 % i == 0):
      return False
  return True
      


def list_coprime(x):
  count = 0
  for i in range(2,x):
    if(coprime_test(x,i)):
      count = count + 1
  print(count)
list_coprime(12)
```

```python
def isTwinPrime(x):
  if(isPrime(x) and isPrime(x+2)):
    return True
  return False

def listTwinprime(x,y):
  for i in range(x,y+1):
    if (isTwinPrime(i)):
      print(i, " ", i+2)
      i += 2

listTwinprime(2,100)
```
