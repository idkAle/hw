```python
import numpy as np

def theMax(numbers):
  my_max = numbers[0]
  for i in numbers:
    if my_max < i:
      my_max = i
  return my_max

def theMin(numbers):
  my_min = numbers[0]
  for i in numbers:
    if my_min > i:
      my_min = i
  return my_min

def theStdDeviation(numbers):
  n = len(numbers)
  mean = sum(numbers) / n
  var = sum((x - mean)**2 for x in numbers) / n
  deviation = var ** 0.5
  return deviation

def theAvg(numbers):
  n = len(numbers)
  total = 0
  for x in numbers:
    total += x
  avg = total / n
  return avg

a = [3,4,6,7,7,9,10]
print(theAvg(a))
```

```python
def evenfib(k):
  first = 1
  second = 1
  total = first + second
  for i in range(k):
    third = first + second
    if (third % 2 == 0):
      total += third
    first = second
    second = third
  return total

print(evenfib(500))
```
